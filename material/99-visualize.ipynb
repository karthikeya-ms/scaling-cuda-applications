```python
import glob, numpy as np, matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
%matplotlib inline

files = sorted(glob.glob("../output/temperature_*.npy"))
frames = [np.load(f, mmap_mode='r') for f in files]

fig, ax = plt.subplots()
im = ax.imshow(frames[0], origin="lower")
plt.colorbar(im, ax=ax)
ax.set_title(files[0])

def update(i):
    im.set_data(frames[i])
    im.set_clim(vmin=float(frames[-1].min()), vmax=float(frames[-1].max()))
    ax.set_title(files[i])
    return (im,)

ani = FuncAnimation(fig, update, frames=len(frames), interval=200, blit=True)
HTML(ani.to_jshtml())
```
